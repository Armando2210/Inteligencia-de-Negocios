**Filtros**

```python
#filtro 1 Los registros de 5 "hostname" diferentes 
# filtro1 = df["host_name"].head()
# filtro1 = df['host_name'][:5]
filtro1 = df.drop_duplicates('host_name').head() 
filtro1
```

```python
#Convertir archivo filtrado a csv
filtro1.to_csv("filtro1_registros_hostname_diferentes .csv")
```

```python
#Filtro 2 Los registros de los host que se hayan unido a Airbnb "antes del año 2019"
filtro2 = df[(df['host_since'] <= '2019')]
filtro2
```

```python
#Convertir archivo filtrado a csv
filtro2.to_csv("filtro2_registros_host_unido_Airbnb_antes_año2019.csv")
```

```python
#Filtro 3 Los registros de los host que responden "a más tardar en 1 día"
filtro3 = df[df['host_response_time'] == 'within a day']
filtro3
```

```python
#Convertir archivo filtrado a csv
filtro3.to_csv("filtro3_registros_host_responden_más tardar_1 día.csv")
```

```python
#Filtro 4 Los registros que tengan "al menos 5 acomodantes"
filtro4 = df[(df['accommodates'] >= 5)]
filtro4
```

```python
#Convertir archivo filtrado a csv
filtro4.to_csv("filtro4_registros_tengan_al menos_5_acomodantes.csv")
```

```python
#Filtro 5 Los registros de los tipos de cuarto: "Entire home" y "Private room"
filtro5 = df[(df['room_type'] == 'Entire home/apt') | (df['room_type'] == 'Private room')]
filtro5
```

```python
#Convertir archivo filtrado a csv
filtro5.to_csv("filtro5_registros_tipos_cuarto.csv")
```

```python
#Filtro 6 Los registros que "a lo mucho tengan 4 camas"
filtro6 = df[df['beds'] <= 4]
filtro6
```

```python
#Convertir archivo filtrado a csv
filtro6.to_csv("filtro6_registros_tengan_al menos_4_camas.csv")
```

```python
#Filtro 7 Los registros que "tengan 1 baño o que sean superhost" pero con la condición de que "todos sean instant_bookable"
filtro7 = df[((df['bathrooms'] == 1) | (df['host_is_superhost'] == 't')) & (df['instant_bookable'] == 't')]
filtro7
```

```python
#Convertir archivo filtrado a csv
filtro7.to_csv("filtro7_registros_tengan_al menos_1 baño o superhost.csv")
```

```python
#Filtro 8 Los registros cuyo "review_scores_cleanliness sea menor de 3 y review_scores_raiting sea mayor de 3"
filtro8 = df[(df["review_scores_cleanliness"] < 3) & (df["review_scores_rating"] > 3)]
filtro8
```

```python
#Convertir archivo filtrado a csv
filtro8.to_csv("filtro8_registros_ review_scores_cleanliness sea menor de 3 y review_scores_raiting sea mayor de 3.csv")
```
